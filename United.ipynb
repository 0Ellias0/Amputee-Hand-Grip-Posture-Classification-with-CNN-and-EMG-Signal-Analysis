```python
import pandas as pd


J = 10
# List of Excel file names
file_names = [rF"C:\Users\hpalm\Downloads\Sesión 2 (1)\Sesión 2\P{J}\Pose_0.xlsx", rF"C:\Users\hpalm\Downloads\Sesión 2 (1)\Sesión 2\P{J}\Pose_1.xlsx", rF"C:\Users\hpalm\Downloads\Sesión 2 (1)\Sesión 2\P{J}\Pose_2.xlsx", rF"C:\Users\hpalm\Downloads\Sesión 2 (1)\Sesión 2\P{J}\Pose_3.xlsx", rF"C:\Users\hpalm\Downloads\Sesión 2 (1)\Sesión 2\P{J}\Pose_4.xlsx", rF"C:\Users\hpalm\Downloads\Sesión 2 (1)\Sesión 2\P{J}\Pose_5.xlsx"]

# Create an empty DataFrame to store the combined data
combined_data = pd.DataFrame()

# Iterate over each file
i = 0
for file_name in file_names:
    # Read the Excel file into a DataFrame, skipping the first row
    data = pd.read_excel(file_name,engine='openpyxl')
    
    # Add a new column with the file name
    data['Clase'] = i
    
    # Append the data to the combined DataFrame
    combined_data = combined_data.append(data, ignore_index=True)
    i+=1

# Save the combined data to a CSV file
combined_data.to_csv('P10S2.csv', index=False)
```
