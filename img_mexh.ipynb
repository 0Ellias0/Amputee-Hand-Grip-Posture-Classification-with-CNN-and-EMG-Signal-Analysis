```python
import numpy as np
import matplotlib.pyplot as plt
import pywt


# Set the parameters for the CWT
wavelet = 'db4'  # Daubechies 4 mother wavelet
scales = np.arange(1, 128)
sampling_frequency = 200  # Hz

# Perform the CWT
coefficients, frequencies = pywt.cwt(data2, scales, wavelet, sampling_period=1/sampling_frequency)

# Plot the CWT image
plt.imshow(np.abs(coefficients), extent=[0, len(data2), np.min(frequencies), np.max(frequencies)], cmap='viridis', aspect='auto')
plt.xlabel('Time (samples)')
plt.ylabel('Frequency (Hz)')
plt.colorbar()
plt.show()
```
